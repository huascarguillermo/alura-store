## **Importando las bases de datos**

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

## **Analisis de datos**

### **Ingreso total por cada tienda**

```python
tiendas = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4,
}

ingresos_totales = {}

for nombre, df in tiendas.items():
    ingresos_totales[nombre] = df['Precio'].sum()

ingresos_totales
```

### **Ventas por categoria**

```python
categorias_por_tienda = {}

for nombre, df in tiendas.items():
    categorias_por_tienda[nombre] = df['Categoría del Producto'].value_counts()
    
categorias_por_tienda
```

### **Valoración media por tienda**

```python
promedio_calificaciones = {}

for nombre, df in tiendas.items():
    promedio_calificaciones[nombre] = df['Calificación'].mean()

promedio_calificaciones
```

### **Productos más vendidos y menos vendidos**

```python
ventas_por_tienda = {}

for nombre, df in tiendas.items():
    ventas_por_tienda[nombre] = df['Producto'].value_counts()

ventas_por_tienda["Tienda 1"].head()
ventas_por_tienda["Tienda 1"].tail()
```

#### **Visualizar productos más vendidos**

```python
import matplotlib.pyplot as plt

for nombre, ventas in ventas_por_tienda.items():
    plt.figure(figsize=(8,4))
    ventas.head(5).plot(kind='bar')
    plt.title(f"{nombre} - Top 5 Productos Más Vendidos")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad Vendida")
    plt.show()
```

#### **Visualizar productos menos vendidos**

```python
for nombre, ventas in ventas_por_tienda.items():
    plt.figure(figsize=(8,4))
    ventas.tail(5).plot(kind='bar')
    plt.title(f"{nombre} - Productos Menos Vendidos")
    plt.xlabel("Producto")
    plt.ylabel("Cantidad Vendida")
    plt.show()
```
